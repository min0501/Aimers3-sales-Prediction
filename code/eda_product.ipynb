#### 가격포지셔닝 컬럼

- 가격포지셔닝 - 같은 대분류 안에서 평균가격바탕으로 표준화

```python
train_data['대분류'].value_counts()
```

- 판매금액 데이터 만들기

```python
info = train_data.iloc[:, :6]
train = train_data.iloc[:, 6:]
sales = sales_data.iloc[:, 6:]
price = sales.div(train)
price1 = pd.concat([info, price], axis=1)
```

```python
price1
```

```python
def mean_excluding_zeros(row):
    non_zero_values = row[row != 0]
    if len(non_zero_values) > 0:
        return non_zero_values.mean()
    else:
        return 0  # Return 0 if all values are zeros

price_means = price.apply(mean_excluding_zeros, axis=1)
```

```python
price_means1 = pd.concat([train_data[['대분류']], price_means],axis=1)
price_means1.columns = ['대분류', '가격평균']
price_means1
```

```python
price_means1[price_means1['가격평균'].isna()]
```

```python
price_means1['가격평균'].fillna(0, inplace=True)
```

- 대분류별 그룹화

```python
brand_stats = price_means1.groupby('대분류')['가격평균'].agg(['mean', 'std']).reset_index()
brand_stats.rename(columns={'mean': 'MeanPrice', 'std': 'StdDevPrice'}, inplace=True)
```

```python
price_means2 = pd.merge(price_means1, brand_stats, on='대분류', how='left')
```

```python
price_means2
```

```python
price_means2['ZScore_대분류'] = (price_means2['가격평균'] - price_means2['MeanPrice']) / price_means2['StdDevPrice']
```

```python
price_means2
```

```python
merged_df5 = pd.concat([merged_df4, price_means2[['가격평균', 'ZScore_대분류']]], axis=1)
merged_df5[merged_df5['브랜드']=='B002-00002']
```

- 취급 제품 수, 언급량, 판매량 저장: lg 생활건강이 취급하는 품목에 대해 특정 브랜드가 차지하는 비율과 연관 있을 것이다.

```python
# 브랜드 별 언급량, 제품수, 판매량 등 저장
brand = keyword_data.iloc[:,:1].copy()
brand['언급량'] = keyword_data.fillna(0).iloc[:,1:].sum(axis=1).to_numpy()
brand['제품수'] = train_data['브랜드'].value_counts().sort_index(ascending=True).to_numpy()
```

```python
train_data_category = train_data.loc[:,:'브랜드'] # 카테고리만 따로 저장
train_data.iloc[:,6:].sum(axis=1).to_numpy()
train_data_category['판매량'] = train_data.iloc[:,6:].sum(axis=1).to_numpy()
grouped = train_data_category['판매량'].groupby(train_data_category['브랜드'])
brand['총판매량'] = grouped.sum().to_numpy()
brand['제품당판매량'] = grouped.mean().to_numpy()
# brand.to_csv('./brand_eda.csv')
```

```python
# 제품 수와 그에 따른 순위를 추가
market_share = brand.iloc[:,[0,2]].sort_values('제품수',ascending=False)
market_share['제품수순위'] = market_share['제품수'].rank(method='min',ascending=False)
market_share
```
