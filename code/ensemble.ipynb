```python
import pandas as pd
import numpy as np

def ReLU(x):
    return np.maximum(0, x)
```

```python
file1 = pd.read_csv('./submit/patchtst_submit_0824.csv').drop(['ID'],axis=1)
file2 = pd.read_csv('./submit/patchtst_submit_0827.csv').drop(['ID'],axis=1)
file3 = pd.read_csv('./submit/patchtst_submit_0828.csv').drop(['ID'],axis=1)
```

```python
file1_R = ReLU(file1)
file2_R = ReLU(file2)
file3_R = ReLU(file3)
```

```python
new_sub = round(0.4*file1_R + 0.3*file2_R + 0.3*file3_R)
```

```python
new_sub
```

```python
submit = pd.read_csv('./dataset/sample_submission.csv')
submit.iloc[:,1:] = new_sub
submit.head()
```
