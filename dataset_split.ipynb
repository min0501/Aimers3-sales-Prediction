### Dataset Split
- 모델을 나누기 위해 데이터셋을 분류하였습니다.

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("./dataset/train.csv")
```

```python
train['쇼핑몰'].value_counts()
```

- 가장 많은 제품을 가진 쇼핑몰 하나와 나머지 쇼핑몰에 대해 다른 모델을 학습시킵니다.

```python
train1 = train[train['쇼핑몰'] == 'S001-00001']
train2 = train[train['쇼핑몰'] != 'S001-00001']
```

```python
train1
```

```python
train1 = train1.drop(columns=['쇼핑몰']) # train1은 같은 쇼핑몰만 포함하였기 때문에 쇼핑몰 컬럼을 삭제합니다.
```

```python
train1.to_csv('./dataset/train1.csv')
train2.to_csv('./dataset/train2.csv')
```
